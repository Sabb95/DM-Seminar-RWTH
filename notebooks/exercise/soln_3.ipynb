```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 13})
import numpy as np
from classy import Class
```

```python
common_settings = {'output':'tCl,pCl,lCl','lensing':'yes'}

DM = Class() 
lcdm = Class()


DM.empty()
lcdm.empty()


DM.set(common_settings)
DM.set({'DM_annihilation_efficiency' : 1.2483e-23})

lcdm.set(common_settings)

DM.compute()
lcdm.compute()
```

```python
l = DM.lensed_cl()['ell'][2:2000]

DM_clTT = DM.lensed_cl()['tt'][2:2000] 
clTT = lcdm.lensed_cl()['tt'][2:2000]

DM_clTE = DM.lensed_cl()['te'][2:2000]
clTE = lcdm.lensed_cl()['te'][2:2000]

DM_clEE = DM.lensed_cl()['ee'][2:2000]
clEE = lcdm.lensed_cl()['ee'][2:2000]
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,4))

ax[0].plot(l,(l*(l+1)*(DM_clTT-clTT))/(2*np.pi), 'r-', label = r'$\mathrm{\delta C_l ^{tot}}$ [ TT ]')
ax[0].set_xlabel(r'$\ell$')

ax[1].plot(l,(l*(l+1)*(DM_clTE-clTE))/(2*np.pi) , label = r'$\mathrm{\delta C_l ^{tot}}$ [ TE ]')
ax[1].set_xlabel(r'$\ell$')

ax[2].plot(l,(l*(l+1)*(DM_clEE-clEE))/(2*np.pi), 'g-',label = r'$\mathrm{\delta C_l ^{tot}}$ [ EE ]')
ax[2].set_xlabel(r'$\ell$')

fig.legend(ncol=3,mode='expand',bbox_to_anchor=(0.,1.1,1.,0.))
plt.tight_layout()
```
