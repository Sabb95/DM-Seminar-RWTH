```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 13})
import numpy as np
from scipy.interpolate import interp1d as interpolater
from classy import Class
```

# $\Lambda$CDM Cosmology 

```python
#Lambda CDM with default standard values
lcdm = Class()

# Choose'halofit' code to generate 'non linear corrections'
# Choose 'lensing' corrections to be 'yes'
# Set a maximum k_max upto which P_m(k) is to be computed
lcdm.set({'output' : 'tCl,lCl,mPk','P_k_max_1/Mpc':3.0 , 'non linear' : 'halofit' ,'lensing' : 'yes'})
lcdm.compute()
```

# Evolution of Background densities

### Get table of background quantities as a python dictionary

```python
background = lcdm.get_background()
background.keys()
```

### Plot densities as a function of redshift

```python
i = 2000
fig,ax = plt.subplots()
plt.rcParams.update({'font.size': 13})
ax.plot(background['z'][i:],background['(.)rho_cdm'][i:], label = r'$\rho_{cdm}$')
ax.plot(background['z'][i:],background['(.)rho_g'][i:], label = r'$\rho_{\gamma}$')
ax.plot(background['z'][i:],background['(.)rho_lambda'][i:], label = r'$\rho_{\Lambda}$')
ax.invert_xaxis()
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('z')
ax.set_ylabel('Densities '+ r'$[\mathrm{Mpc}]^{-3}$')
plt.legend()
plt.tight_layout()
fig.savefig('densities.png',dpi=300)
```

# $C_\ell$

## Lensed vs Unlensed $C_\ell$ spectrum 

```python
# Choose from l >=2 neglecting the monopole and dipole terms

l = lcdm.lensed_cl()['ell'][2:]
clTT_len = lcdm.lensed_cl()['tt'][2:]
clTT_raw = lcdm.raw_cl()['tt'][2:2501]

fig, ax = plt.subplots()
ax.plot(l,l*(l+1)*clTT_raw, label = 'Lensed')
ax.plot(l,l*(l+1)*clTT_len, label = 'Unlensed')
ax.set_xlabel(r'$\ell$')
ax.set_ylabel(r'$ \ell (\ell + 1) C_\ell$')
ax.legend()
ax.set_xscale('log')
fig.savefig('clTT_lens_and_raw.png',dpi=300)
```

## $C_\ell$ spectrum for $\Lambda$CDM 

```python
l = lcdm.lensed_cl()['ell'][2:]
clTT_len = lcdm.lensed_cl()['tt'][2:]
clTT_raw = lcdm.raw_cl()['tt'][2:2501]
fig, ax = plt.subplots()
ax.plot(l,l*(l+1)*clTT_raw, 'r-')
ax.set_xlabel(r'$\ell$')
ax.set_ylabel(r'$ \ell (\ell + 1) C_\ell$')
ax.set_xscale('log')
fig.savefig('clTT_example_slide.png',dpi=300)
```

## $C_\ell$ without lensing correction for varying (CDM+$\Lambda$)

```python
varying_cdm = Class()
varying_cdm.set({'output' : 'tCl'})
cl = lcdm.raw_cl()['tt'][2:]
l = lcdm.raw_cl()['ell'][2:]
lam_lcdm = lcdm.Omega_Lambda()

omega_c_list = [0.40,0.2,0.1]
omega_max = np.amax(omega_c_list)

plt.plot(l,l*(l+1)*cl,label=r'$\Lambda$CDM')
for i in omega_c_list :
    varying_cdm.set({'Omega_cdm' : i, 'output':'tCl'})
    varying_cdm.compute()
    cl = varying_cdm.raw_cl()['tt'][2:]
    l = varying_cdm.raw_cl()['ell'][2:]
    plt.plot(l,l*(l+1)*cl,'r-',alpha=i/omega_max,label= r'$\Omega_c = {:.2f}, \Omega_\Lambda = {:.2f} $'.format(i,varying_cdm.Omega_Lambda())      )

plt.xscale('log')
plt.legend()
plt.savefig('varying_cdm.png', dpi=300)
```

# Matter power spectrum $P_m(k)$

## $P_m(k)$ for varying (Baryons + CDM)

```python
varying_b = Class()
blist = [0.005,0.02,0.05,0.10]
kk = np.logspace(-4,np.log10(3),1000)
plt.rcParams.update({'font.size': 13})
for b in blist :
    c = 0.35 - b 
    
# If you have extreme values of Omega_b it is always better to specify the 'YHe to avoid warnings and error'
    varying_b.set({'Omega_b' : b, 'Omega_cdm' : c ,'P_k_max_1/Mpc': 3.0 ,'YHe' : 0.24})
    varying_b.set({'output' : 'mPk'})
    varying_b.compute()
    h = varying_b.h()
    Pk = []
    if b==0.05 : 
        col = 'r-'
    else :
        col = 'black'
    for k in kk :
        Pk.append(varying_b.pk(k*h,0.)*h**3)
    plt.xscale('log');plt.yscale('log');plt.xlim(kk[0],kk[-1])
    plt.xlabel(r'$\mathrm{k} \,\,\,\, ( \mathrm{h} \ \mathrm{Mpc}^{-1})$')
    plt.ylabel(r'$\mathrm{P(k)} \,\,\,\, (\mathrm{h}^{-3} \mathrm{Mpc}^3 )$')
    plt.plot(kk,Pk, col, label = r'$\Omega_b$ = {:.2f}'.format(b) ) 
    plt.xlim(5*1e-3,1.25)
    plt.title(r'Varying Matter composition for $\Omega_m$ = 0.35')
    plt.ylim(1e2)
plt.tight_layout()
plt.savefig('bao.png',dpi=300)
```

## Non-Linear Matter power spectrum

```python
# plot P(k) linear/nonlinear at redhsift z=0 and z=2
kk = np.logspace(-4,np.log10(3),1000) # k in h/Mpc
Pk_L_0 = []  # P_L(k,0) in (Mpc/h)**3
Pk_NL_0 = [] # P_NL(k,0) in (Mpc/h)**3
h = lcdm.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk:
    Pk_L_0.append(lcdm.pk_lin(k*h,0.)*h**3) 
    Pk_NL_0.append(lcdm.pk(k*h,0.)*h**3) 
    
# plot P(k)
plt.figure(3)
plt.xscale('log');plt.yscale('log');plt.xlim(kk[0],kk[-1])
plt.xlabel(r'$\mathrm{k} \,\,\,\, [\mathrm{h} \ \mathrm{Mpc^{-1}}]$')
plt.ylabel(r'$\mathrm{P(k)} \,\,\,\, [\mathrm{Mpc}^3 \mathrm{h}^{-3}]$')
plt.plot(kk,Pk_L_0,':',label='Linear')
plt.plot(kk,Pk_NL_0,'-',label = 'Non Linear')
plt.legend()
plt.savefig('nonlinear.png',dpi=300)
```
